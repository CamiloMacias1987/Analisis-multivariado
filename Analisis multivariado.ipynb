# Ejemplo 2

````python
#Importamos las librerias
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
````

````python
url='https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
df =pd.read_csv(url,sep=',')
df.columns= ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome']
df.head()
````

*   **Pregnancies (Embarazos):** Esta variable podría representar el número de embarazos que ha tenido una persona. En el contexto de datos médicos, podría usarse para registrar la historia de embarazos de una paciente.

*   **Glucose (Glucosa):** La glucosa es un tipo de azúcar en la sangre y su concentración se mide en miligramos por decilitro (mg/dL). Esta variable podría indicar el nivel de glucosa en la sangre de una persona. Es importante para el diagnóstico y el monitoreo de la diabetes.

*   **BloodPressure (Presión Sanguínea):** Esta variable podría representar la presión sanguínea de una persona en milímetros de mercurio (mm Hg). La presión sanguínea se mide en dos valores: la presión sistólica (la presión cuando el corazón late) y la presión diastólica (la presión cuando el corazón está en reposo).

*   **SkinThickness (Espesor de la Piel):** Esta variable podría representar el espesor de la piel en algún lugar del cuerpo. Podría ser relevante en estudios médicos o dermatológicos.

*   **Insulin (Insulina):** La insulina es una hormona producida por el páncreas que regula los niveles de glucosa en sangre. La medición de insulina podría ser relevante para evaluar la función del páncreas y el tratamiento de la diabetes.

*   **BMI (Índice de Masa Corporal):** El índice de masa corporal es una medida que relaciona el peso y la altura de una persona y se utiliza para evaluar el peso corporal. Puede ser útil para determinar si una persona tiene un peso saludable, bajo peso, sobrepeso u obesidad.

*   **DiabetesPedigreeFunction (Función de Pedigrí de Diabetes):** Esta variable podría representar una función que evalúa la susceptibilidad genética a la diabetes. Puede incluir información sobre la historia familiar de diabetes.

*   **Age (Edad):** La edad de una persona, medida en años. Es una variable importante en muchos tipos de análisis, ya que la edad puede estar relacionada con diversas condiciones médicas.

*   **Outcome:** Esta variable podría indicar el resultado de un diagnóstico o prueba, como si una persona tiene diabetes o no (por ejemplo, 1 para diabetes, 0 para no diabetes).

````python
#Reemplazar el Outcome de 1 a 'Diab' y 0 a 'Non-Diab'
df.Outcome = df.Outcome.replace({0:'Non-Diab',1:'Diab'})
df.head()
````

````python
df.shape
df.info()
````

````python
df.describe().T
````

### **Pairplot**

````python
plt.figure(dpi = 120) #NItidez pixels
sns.pairplot(df, hue='Outcome', palette='plasma')
plt.legend(['Non Diabetic','Diabetic']) #legenda
plt.show()
````

### **Gráfico de correlacion**

````python
plt.figure(dpi = 120,figsize= (5,4))
mask = np.triu(np.ones_like(df.corr(),dtype = bool))
sns.heatmap(df.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
plt.yticks(rotation = 0)
plt.xticks(rotation = 90)
plt.title('Correlation Heatmap')
plt.show()
````

### **Jointplot**

La función jointplot en Seaborn se utiliza para crear una visualización que muestra la relación entre dos variables, así como las distribuciones marginales de cada variable. Aquí está la sintaxis básica de jointplot:

```
sns.jointplot(x='variable_x', y='variable_y', data=df, kind='scatter')
plt.show()
```
*   **x:** El nombre de la variable que se representará en el eje x del gráfico.
*   **y:** El nombre de la variable que se representará en el eje y del gráfico.
*   **data:** El DataFrame que contiene los datos.
*   **kind:** El tipo de gráfico que se desea. Puede ser 'scatter' para un gráfico de dispersión, 'kde' para un gráfico de densidad kernel, 'hist' para un histograma, entre otros.






````python
plt.figure(dpi = 100, figsize = (5,4))
print("Joint plot de Glucosa con otras variables ==> \n")
for i in  df.columns:
    if i != 'Glucose' and i != 'Outcome':
        print(f"Correlacion entre Glucose y {i} ==> ",df.corr().loc['Glucose'][i])
        sns.jointplot(x='Glucose',y=i,data=df,kind = 'scatter',color = 'purple')
        plt.show()
````

````python
#Jointplot presión sanguínea y otras variables
col = list(df.columns)
idx = col.index('BloodPressure')

plt.figure(dpi = 100, figsize = (5,4))
print("Joint plot de BloodPressure con otras variables ==> \n")
for i in  range(idx+1,len(col)-1):
    print(f"Correlacion entre BloodPressure y {col[i]} ==> ",df.corr().loc['BloodPressure'][col[i]])
    sns.jointplot(x='BloodPressure',y=col[i],data=df,kind = 'scatter',color = 'green')
    plt.show()
````

````python
#Espesor de la piel
col = list(df.columns)
idx = col.index('SkinThickness')

plt.figure(dpi = 100, figsize = (5,4))
print("Joint plot de SkinThickness con otras variables ==> \n")
for i in  range(idx+1,len(col)-1):
    print(f"Correlacion entre SkinThickness y {col[i]} ==> ",df.corr().loc['SkinThickness'][col[i]])
    sns.jointplot(x='SkinThickness',y=col[i],data=df,kind = 'reg',color = 'blue')
    plt.show()
````
